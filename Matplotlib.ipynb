```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

data_one = np.arange(100,201)
data_one
```

```python
plt.plot(data_one)
plt.show()
```

```python
# 通过figure（）函数创建画布
figure_obj = plt.figure()
data_two = np.arange(200,301)
plt.figure(facecolor='gray')
plt.plot(data_two)
plt.show()
```

```python
# subplot() 函数创建单个子图
nums = np.arange(0,101)
plt.subplot(221) # 生成2*2的矩阵区域，占用编号为1的区域
plt.plot(nums, nums)
plt.subplot(222)
plt.plot(nums, -nums)
plt.subplot(212)
plt.plot(nums, nums**2)
plt.show()
```

```python
# 通过subplots()函数创建多个子图
nums = np.arange(1,101)
fig, axes = plt.subplots(2,2)  # 分成2*2的矩阵区域，返回子图数组axes
ax1 = axes[0,0]
ax2 = axes[0,1]
ax3 = axes[1,0]
ax4 = axes[1,1]
ax1.plot(nums, nums)
ax2.plot(nums, -nums)
ax3.plot(nums, nums **2)
ax4.plot(nums, np.log(nums))
plt.show()
```

```python
# 通过add_subplot()函数添加和选中子图
fig = plt.figure()
fig.add_subplot(2,2,1)
fig.add_subplot(2,2,2)
fig.add_subplot(2,2,4)
fig.add_subplot(2,2,3)
random_arr = np.random.randn(100)
plt.plot(random_arr) # 默认是在最后一次使用的位置上作图
plt.show()
```

```python
data = np.arange(0,1.1,0.01)
plt.title('Title')
plt.xlabel("x")
plt.ylabel('y')
# 设置x/y轴的刻度
plt.xticks([0,0.5,1])
plt.yticks([0,0.5,1.0])
plt.plot(data, data**2)
plt.plot(data, data**3)
plt.legend(["y=x^2","y=x^3"]) # 添加图例
plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x + 1
y2 = x**2
# num=3 代表图片上方标题变为figure3，figsize=(长，宽)
plt.figure(num=3,figsize=(8,5))
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')
plt.xlim((-1,2)) # x轴范围
plt.ylim((-2,3))
# 英文直接写，中文需要加上fontproperties属性
plt.xlabel(u'价格',fontproperties='SimHei',fontsize=16)
plt.ylabel(u'利润',fontproperties='SimHei',fontsize=16)
new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)
'''
设置对应坐标用汉字或英文表示，后面的属性fontproperties表示中文可见，不乱码，
内部英文$$表示将英文括起来，r表示正则匹配，通过这个方式将其变为好看的字体
如果要显示特殊字符，比如阿尔法，则用转意符\alpha,前面的\ 表示空格转意
'''
plt.yticks([-2,-1.8,-1,1.22,3.],
          ['非常糟糕','糟糕',r'$good\ \alpha$',r'$really\ good$','超级好'],
          fontproperties='SimHei',
          fontsize=12,
          fontset='custom')
plt.show()
```

```python
arr_random = np.random.randn(100)
# 直方图
plt.hist(arr_random, bins=8, color='g', alpha=0.7)
plt.show()
```

```python
x = np.arange(51)
y = np.random.randn(51)*10
# 绘制散点图
plt.scatter(x,y)
plt.show()
```

```python
x = np.arange(5)
# 从上下限范围随机取整，创建两个2行5列的数组
y1,y2 = np.random.randint(1,31,size=(2,5))
width = 0.25
ax = plt.subplot(1,1,1)
ax.bar(x,y1,width,color='r')
ax.bar(x+width,y2,width,color='g')
ax.set_xticks(x+width)
ax.set_xticklabels(['January','February','March','April','May'])
plt.show()
```

```python
data = np.arange(1,3,0.3)
# 绘制直线，颜色：青色，标记：x，线型：长虚线
plt.plot(data, color="c",marker="x",linestyle="--")
# 绘制直线，颜色：品红色，标记：实心圆，线型：短虚线
plt.plot(data+1, color="m", marker="o",linestyle=":")
# 绘制直线，颜色：黑色，标记：五边形，线型：短点相间线
plt.plot(data+2, color="k", marker="p",linestyle="-.")
# 也可以采用下面的方式绘制三条不同颜色，标记和线型的直线
#plt.plot(data,'cx--',data+1,'mo:',data+2,'kp-.')
plt.show()
```
